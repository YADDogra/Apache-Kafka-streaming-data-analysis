# Kafka Producer

## 1.2 Memory Event Producer

```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv


def read_csv(fileName):
    input_file = csv.DictReader(open(fileName))
    data = [i for i in input_file]
    return data

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    # assign the topic name
    topic = 'Memory_Stream'
    
    producer = connect_kafka_producer()
    
    # read the data
    data_all = read_csv('Streaming_Linux_memory.csv')
    print('Publishing records..')
    
    #initialize counter
    step_1 = 0 
    step_2 = 0
    
    while True:
        
        # random number of rows for first batch
        send_frow = random.randint(20,80)
        
        # random number of rows for second batch
        send_srow = random.randint(0,5)
        
        # condition to check the length of the data
        # is not violated for both the batches
        if ((step_1 + send_frow) <= len(data_all)) and ((step_2 + send_srow) <= len(data_all)):
            
            fb_send = data_all[step_1:step_1+send_frow] 
            sb_send = data_all[step_2:step_2+send_srow]
            
        else:
            step_1 = 0
            step_2 = 0
            fb_send = data_all[step_1:step_1+send_frow] 
            sb_send = data_all[step_2:step_2+send_srow]
            
        # add the timestamp column
        for t in fb_send:
            t['ts'] = dt.datetime.utcnow().strftime("%s")
        
        # add the timestamp column
        for t in sb_send:
            t['ts'] = dt.datetime.utcnow().strftime("%s")
        
        # increment counter
        step_1 = step_1 + send_frow
        step_2 = step_2 + send_srow
            
        # publish the first batch and sleep for 10 sec to publish other
        publish_message(producer, topic, fb_send)
        print('--------------------------------------')
        sleep(10)    
        publish_message(producer, topic, sb_send)
              
```
